# Impact du nombre de Pas √† nombre d'Epoch Fixe

```python
StepsImpact = []
StepsCount = range(50,2050,100)
for Steps in StepsCount:
    print(Steps)
    policyLosses, valueLosses, rewardsTotal = main.run(epochs = 400, steps = Steps, lr = 3e-4, minibatch = 64, epsilon =0.2)
    StepsImpact.append(sum(rewardsTotal[375:])/25*(2000/Steps))#Average on 25 last steps and normalized to have 2000 steps Reward
    print (sum(rewardsTotal[375:])/25*2000/Steps)
```

```python
plt.plot(StepsCount,StepsImpact)
```

# Impact de Epsilon

```python
epsilonImpact = []
epsilonCount = [0.1+k*0.1 for k in range(10)]
for epsilon in epsilonCount:
    policyLosses, valueLosses, rewardsTotal = main.run(epochs = 200, steps = 2000, lr = 3e-4, minibatch = 64, epsilon=epsilon)    
    epsilonImpact.append(sum(rewardsTotal[190:])/10)
    print(epsilon)
```

```python
plt.plot(epsilonCount,epsilonImpact)
```
